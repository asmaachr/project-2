```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from scipy import stats
```

```python
# Load Bikes to dataframe
city_bikes_df = pd.read_csv('bikes')

# Load Yelp1 to dataframe
Yelp_df=pd.read_csv('Yelp1')

# Load Foursquare to dataframe
FSQ_df=pd.read_csv('Foursquare1')

city_bikes_df.head()
```

```python

# creating an id based on index -station as it was created in previous table i= referto long, lat of the row
city_bikes_df['station']= city_bikes_df.index
# dropping column stationid

city_bikes_df=city_bikes_df.drop('stationid', axis=1)

city_bikes_df
```

```python
# Yelp_df.isnull().sum()
parks_yelp_df=Yelp_df[Yelp_df['category']=='park']
restaurants_yelp_df=Yelp_df[Yelp_df['category']=='restaurant']
#Yelp Park summary
agg_dict={'rating':'mean', 'category':'count'}
parks_yelp_df=parks_yelp_df.groupby(by='station', as_index=False).agg(agg_dict)
parks_yelp_df = parks_yelp_df.rename(columns={'category': 'Yelp_park_count'})

parks_yelp_df = parks_yelp_df.rename(columns={'rating': 'Yelp_park_rating'})
parks_yelp_df

# #Yelp Park summary
agg_dict={'rating':'mean', 'category':'count'}
restaurants_yelp_df=restaurants_yelp_df.groupby(by='station', as_index=False).agg(agg_dict)

restaurants_yelp_df = restaurants_yelp_df.rename(columns={'category': 'Yelp_restaurant_count'})

restaurants_df = restaurants_yelp_df.rename(columns={'rating': 'Yelp_restaurant rating'})



yelp_summary = pd.merge(restaurants_df, parks_yelp_df, on='station', how='outer')

yelp_summary = yelp_summary.fillna(0)
yelp_summary.describe()

```

```python
FSQ_df.isnull().sum()
```

```python
FSQ_df['POI'].value_counts()
```

```python
# Soting the FSQ data by category

parks_FSQ_df=FSQ_df[FSQ_df['category']=='park']
restaurants_FSQ_df=FSQ_df[FSQ_df['category']=='restaurant']

##FSQ Park summary

agg_dict={'category':'count'}
parks_FSQ_df = parks_FSQ_df.groupby(by='station', as_index=False).agg(agg_dict)
parks_FSQ_df = parks_FSQ_df.rename(columns={'category': 'FSQ_park_count'})


##FSQ restaurant summary
agg_dict={'category':'count'}
restaurants_FSQ_df=restaurants_FSQ_df.groupby(by='station', as_index=False).agg(agg_dict)

restaurants_FSQ_df = restaurants_FSQ_df.rename(columns={'category': 'FSQ_restaurant_count'})

## FSQ summary
FSQ_summary = pd.merge(restaurants_FSQ_df, parks_FSQ_df, on='station', how='outer')
## Filling NaN with 0: if a POI is NaN mean count =0 & for rating assuming not good
FSQ_summary = FSQ_summary.fillna(0)
FSQ_summary.describe()

```

```python
# joining the 3 dfs
### Note: keeping the scope of the project in mind I prefered to join the summary dfs wit the bikes one.

bikes_yelp_data = pd.merge(city_bikes_df, yelp_summary, on='station', how='left')
bikes_yelp_fsq_data = pd.merge(bikes_yelp_data, FSQ_summary, on='station', how='left')

bikes_yelp_fsq_data.head(100)
```

```python
bikes_yelp_fsq_data.describe()
```

```python
bikes_yelp_fsq_data.isnull().sum()
```

```python
# two stations have no data in Yelp
## looking to the data_ yelp shows many big parks where FSQ shows parks and also some places that not considered park
bikes_yelp_fsq_data.fillna(0, inplace=True)
bikes_yelp_fsq_data.replace([np.inf, -np.inf], 0, inplace=True)
bikes_yelp_fsq_data.isnull().sum()
```

```python
bikes_yelp_fsq_data.info()
```

```python
bikes_yelp_fsq_data = bikes_yelp_fsq_data.astype({'Yelp_restaurant_count': int, 'Yelp_park_count': int, 'FSQ_restaurant_count': int,'FSQ_park_count':int})
```

```python
df=bikes_yelp_fsq_data
analysis_df=df.drop(['name','longitude', 'empty_slots','station'], axis='columns')
analysis_df.rename(columns={'latitude': 'lat'})
print(analysis_df.columns)
analysis_df= analysis_df.drop('latitude ', axis= 'columns')
analysis_df['perc_of_free_bikes%']=analysis_df['free_bikes']*100/analysis_df['total_bikes']

analysis_df['perc_of_free_bikes%']=np.round(analysis_df['perc_of_free_bikes%'],2)
analysis_df.head()

```

```python
sns.pairplot(analysis_df)
plt.show()
```

### Analysis
- Free_bikes: distribution is not clear -maybe we need to review the bins with
- Total bikes by station: seems close to normal right skewed
- yelp restaurant rating: shows that the mode is 5 for restaurants which meaning that most restaurants as rated 5
- yelp restaurant count: almost all station have more than 20 restaurants around within radius 1000
- yelp park average rating: we will ignore the first bar as it is refering to stations without park. the average rating of the parks is around 1 (suprising compared to google)
- yelp park count: again here we can see that yelp is pulling only big major parks
- FSQ restaurant count: almost all stations have about 10 restaurants around within radius 1000
- FSQ park count: mode=10, majority of station have 10 parks around within radius 1000
- per_free_bikes: not clear maybe we need to redo it by changing bins or testing the normality of the distribution
- there is no obvious correlation between parameters

```python
# correlation matrix

corrmat = analysis_df.corr()
sns.heatmap(corrmat, vmax=.8, square=True, cmap="RdYlGn_r")
plt.show()
```

There is no obivious correlation between the variables except Yelp count/rating because used the mean and perc_of_free_bikes and free_bikes wich is normal


```python
# testing the total bikes, free_bikes for, per_free_bikes normality

stat, p = stats.shapiro(analysis_df['total_bikes'])
print('%0.15f' % p)
```

```python
# testing the total bikes, free_bikes for, per_free_bikes normality

stat, p = stats.shapiro(analysis_df['free_bikes'])
print('%0.15f' % p)
```

```python

stat, p = stats.shapiro(analysis_df['perc_of_free_bikes%'])
print('%0.15f' % p)
```

The three distributions are NOT normal according to the test since all Pvalues<0.05 therefore we  reject H0 that the data is normally distributed

```python
# comparaison between FSQ and Yelp regarding the count of restaurants by station
#- the parks is obvious

fig, ax = plt.subplots()
ax.boxplot([df['FSQ_restaurant_count'], df['Yelp_restaurant_count']])
ax.set_xticklabels(['FSQ_restaurant_count', 'Yelp_restaurant_count'])
ax.set_ylabel('value')
plt.show()
```

Yelp definetly provide more results by station in Hamilton
We can confirm with a ttest

```python
stat, p = stats.ttest_ind(df['FSQ_restaurant_count'], df['Yelp_restaurant_count'])
p
```

Since our Value is < 0.0.5 we reject H0 therefore the two averages are differents
