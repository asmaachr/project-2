Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python

# Create a fuction to send the request with variables 

def getdata(latitude, longitude,x):   
    ## create the API request URL
    url = 'https://api.foursquare.com/v3/places/search'
    api_key = os.environ['FOURSQUARE_API_KEY']
    
    
    params = {
        'll': f'{latitude},{longitude}', 'radius':'1000', 'query':x
        
    }
        # Create dictionary for headers
    headers = {"Accept": "application/json"}
    # Add key with our API KEY
    headers['Authorization'] = api_key
    
    
    ## make the GET request
    
    response = requests.get(url, params=params, headers=headers)
    
    # Get the JSON content from the response
    
    results = response.json()
    
  
    ## Print out the entire JSON response to see catergory names
    results_list = results['results']
    
    ## Partial parsing by POI object
    ### Keeping the scope of the project in mind, I decided to get only the name of the POI(restaurant or Park). Foursquare doesn't display ratings/or reviews count
    ### looping on the size of the request = station
    ### Note: I notice that when I pull the query with two parameters, It is hard to identify a sort item by POI type. 
    ###I decided to request information by type of POI=x
    
    size=len(results_list)
     
    x_name_list=[]
        
    for i in range(size):
        # name=results_list[i]['name']
        
        x_name_list.append(results_list[i]['name'])
            
    return x_name_list
              
```

```python
getdata(43.259125842121065,-79.87721174955368,'Restaurant')
```

```python

# Parsing for one station i


def alldata (latitude, longitude):
    
  
    response=[]
    station_data=[]
    
## creating a dictionary withrequest and creating a category for POI 
###Looping on every POI for Parks   
    
    response= getdata(latitude, longitude,'Park')   


    for i in range(len(response)):
        
        objet={}
        objet['category']= 'park'
        objet['latitude']= latitude
        objet['longitude']= longitude
        objet['POI']= response[i]
        station_data.append(objet)
        
###Looping on every POI for restaurants      
    response= getdata(latitude, longitude,'restaurants')
    for i in range(len(response)):
        
        objet={}
        objet['category']= 'restaurant'
        objet['latitude']= latitude
        objet['longitude']= longitude
        objet['POI']= response[i]
        station_data.append(objet)
### creating a df for requested station
    df=pd.DataFrame(station_data)
    
    return df
```

```python
#testing the function
alldata(43.259125842121065,-79.87721174955368)
```

```python
# read city_bikes csv#
import pandas as pd
city_bikes_df = pd.read_csv('bikes')
city_bikes_df
```

```python
# making sure that the upload happened properly
len(city_bikes_df)
   
```

Put your parsed results into a DataFrame

```python
# creating a dataframe for all Foursquare data

Foursquare_data_dict=[]

x=len(city_bikes_df)

## reading every row of the City bikes and creating its own df
for i in range(x):
    longitude = city_bikes_df.loc[i][2]
    latitude = city_bikes_df.loc[i][3]
   
    temp_data= alldata(latitude,longitude)
    temp_data['station']= i
    Foursquare_data_dict.append(temp_data)

## Concatenating all dfs in one

Foursquare_df=pd.DataFrame()
for df in Foursquare_data_dict:
    x = pd.concat([Foursquare_df, df])
    Foursquare_df= x
    

Foursquare_df.head()

# collecting the data as CSV file
file_path = Path('data/Foursquare1.csv') 
file_path.parent.mkdir(parents=True, exist_ok=True)

Foursquare_df.to_csv(file_path, sep=',',index=False)

```

# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
#Same Process

def getdata1(latitude, longitude,x):   
       
    url = 'https://api.yelp.com/v3/businesses/search'
    api_key = os.environ['YELP_API']
    
 
    
    params = {
        'latitude':latitude,
        'longitude':longitude,
        'radius':'1000',
        'term':x
      
    }
        # Create dictionary for headers
    headers = {"Accept": "application/json"}
    # Add key with our API KEY
    headers['Authorization'] = api_key

    
    
    # make the GET request
    
    response = requests.get(url, params=params, headers=headers)
    
    # Get the JSON content from the response
    
    results = response.json()
    
    results_list = results['businesses']
    
    size=len(results_list)
    
    
    x_name_list=[]
    x_rating_list=[]
    x_review_count_list=[]   
   # looping on every POI 
    
    for i in range(size):
        x_name_list.append(results_list[i]['name'])
        x_rating_list.append(results_list[i]['rating'])
        x_review_count_list.append(results_list[i]['review_count'])    
    return x_name_list,x_rating_list,x_review_count_list
```

```python
# testing the function
getdata1(43.259125842121065,-79.87721174955368, 'restaurant')
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
# function To loop the city_bike file and create df by station

def regroup_data_yelp(latitude, longitude):
    
    response=[]
    station_data=[]
    
    response= getdata1(latitude, longitude,'Park')
    
    for i in range(len(response[0])):
        
        objet={}
        objet['category']= 'park'
        objet['latitude']= latitude
        objet['longitude']= longitude
        objet['POI']= response[0][i]
        objet['rating']= response[1][i]
        objet['reviews_count']=response[2][i]
        station_data.append(objet)
    
    response= getdata1(latitude, longitude,'restaurants')
    for i in range(len(response[0])):
        
        objet={}
        objet['category']= 'restaurant'
        objet['latitude']= latitude
        objet['longitude']= longitude
        objet['POI']= response[0][i]
        objet['rating']= response[1][i]
        objet['reviews_count']=response[2][i]
        station_data.append(objet)

    return pd.DataFrame(station_data)
```

```python
regroup_data_yelp(43.259125842121065,-79.87721174955368)
```

Put your parsed results into a DataFrame

```python
# looping on every station and creating dfis
yelp_data_dict=[]

x=len(city_bikes_df)

for i in range(x):
    longitude = city_bikes_df.loc[i][2]
    latitude = city_bikes_df.loc[i][3]
   
    temp_data= regroup_data_yelp(latitude,longitude)
    temp_data['station']= i
    yelp_data_dict.append(temp_data)

# concatenating dfis

Yelp_df=pd.DataFrame()
for df in yelp_data_dict:
    x = pd.concat([Yelp_df, df])
    Yelp_df = x
    
Yelp_df.head()

#SaveCSV
file_path = Path('data/Yelp1.csv') 
file_path.parent.mkdir(parents=True, exist_ok=True)

Yelp_df.to_csv(file_path, sep=',',index=False)
```
